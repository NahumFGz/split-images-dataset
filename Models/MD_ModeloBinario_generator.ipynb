### Entrenar modelo

```
import numpy as np
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img

def visualize_augmentations(image_path, transformation_ratio=0.2, num_images=5):
    # Cargar la imagen
    img = load_img(image_path)
    x = img_to_array(img)  # Convertir la imagen a un array numpy
    x = np.expand_dims(x, axis=0)  # Añadir una dimensión extra para que sea compatible con ImageDataGenerator

    # Definir el generador de datos con las transformaciones deseadas
    datagen = ImageDataGenerator(
        rescale=1. / 255,
        rotation_range=transformation_ratio,
        zoom_range=transformation_ratio,
        horizontal_flip=True,
        vertical_flip=True
    )

    # Generar y mostrar las imágenes transformadas
    fig, ax = plt.subplots(1, num_images + 1, figsize=(20, 20))

    # Mostrar la imagen original
    ax[0].imshow(img)
    ax[0].set_title("Original")
    ax[0].axis('off')

    # Generar imágenes transformadas y mostrarlas
    img_generator = datagen.flow(x, batch_size=1)

    for i in range(1, num_images + 1):
        batch = next(img_generator)
        img_array = batch[0]  # Obtener la imagen del batch
        ax[i].imshow(img_array)
        ax[i].set_title(f"Augmented {i}")
        ax[i].axis('off')

    plt.tight_layout()
    plt.show()

# Ruta de la imagen a cargar
image_path = '/content/drive/MyDrive/Colab Notebooks/DataSets/Fundus/RIM/Fold0/train/glaucoma/Im001_g_RIM.jpg'
visualize_augmentations(image_path)
```
